Tokenization, stemming and lemmatization with NLTK

Imports

```python
import pandas as pd
import nltk
```

Downloads additional content for the nltk package. 

```python
#nltk.download()
```

Import kaggle dataset. 

```python
csv = pd.read_csv("Musical_instruments_reviews.csv", index_col=False)
```

Confirm import worked. info() also displays useful information about the dataset.

```python
csv.info()
```

Split the summary column from the rest of the dataset.

```python
summary = csv['summary']
```

Check the summary split worked.

```python
print(summary)
```

Change print options to verify the algorithms are running against the entire dataset.

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

Tokenize the summary dataset using the nltk word_tokenize function.

```python
word_tokens = nltk.word_tokenize(str(summary))
```

Verify word tokenization

```python
print(word_tokens)
```

Perform Stemming

```python
from nltk.stem import SnowballStemmer    
```

```python
stemming_array = []
for token in word_tokens:
    stemming_array.append(SnowballStemmer('english').stem(token))
```

Verify stemming

```python
print(stemming_array)
```

Preform lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lem_array = []
for token in word_tokens:
    lem_array.append(lemmatizer.lemmatize(token))
```

Verify lemmatization

```python
print(lem_array)
```
